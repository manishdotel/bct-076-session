```python
import matplotlib.pyplot as plt
x = []
n = range(-10,10,1)
print(len(n))
for i in range (len(n)):
  temp= (0 if n[i]<0 else 1)
  x.append(temp)
plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('n(n)')
plt.title('unit step')
```

```python
import matplotlib.pyplot as plt

n = range(-10,10,1)
a =[2 ,-2 ,0.5 ,0.5]
b = 1
for i in a:
  x = []
  for val in n:
    tmp = pow(i ,val)
    x.append(tmp)

  plt.subplot(2,2,b)
  b+=1
  plt.stem(n,x)
  plt.xlabel('n')
  plt.ylabel('n(n)')
plt.title('a^n')
```
